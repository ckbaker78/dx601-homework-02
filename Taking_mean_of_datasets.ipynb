```python
import numpy as np 
```

```python
data = [3,4,5,7,2,10]
```

```python
sum(data) / len(data) # this is the python method 
```

```python
np.array(data)

np.mean(data)
```

```python
np.std(data)
```

```python
np.var(data)
```

```python
mean = sum(data) / len(data)

population_variance = sum((x - mean)**2 for x in data ) / len(data)

sample_variance = sum((x - mean)**2 for x in data )/ (len(data) - 1) # the n minus one is the sample mean correction 

np.var(data, ddof = 1) # this is the numpy version for adjusted SAMPLE variance (n-1)

standard_deviation = np.sqrt(population_variance)

standard_deviation
```
