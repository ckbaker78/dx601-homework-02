```python
# To calculate Z- Score 

## The Z score tells us how far away a particular value is away from the mean of a dataset
### The Z score = Selected Value - mean of dataset / standard deviation of the dataset

list2 = [100, 200, 300, 400, 500]

list1 =[1, 2, 3, 4, 5]




Zscore_list1 = (5 - np.mean(list1))/ np.std(list1)

Zscore_list2 = (500 - np.mean(list2))/ np.std(list2)

#### note that the Z-score of these two list of numbers is equal, meaning the selected last value lie equal standard deviations away from the mean of the dataset ####

print(Zscore_list1,Zscore_list2)
```

```python
# Bootstraping a dataset - this is used to approximate distribution of sample estimates when sampling the whole population 

import random 


resampled_means = [] # initialize an empty array 




for i in range(1000): 
    resampled_means.append(np.mean(random.choices(data, k = len(data))))




resampled_means[:10]
```

```python
np.mean(resampled_means) # the mean of the resampled data gives you a much closer value to the mean of the dataset.
```
